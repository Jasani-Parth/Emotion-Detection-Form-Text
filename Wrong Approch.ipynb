```python
def Word_Frequency_Cal(label) :
    
    # first gets that all data_frame
    temp_df = df1[df1.label==label]
    
    # extract texts from new data_frame
    messages = np.array(temp_df['text'])
    
    Freq = dict()
    
    for m in messages :
        for word in m.split(' ') :
            if word in list(Freq.keys()) :
                Freq[word] += 1
            else :
                Freq[word] = 1
    
    return Freq
```
